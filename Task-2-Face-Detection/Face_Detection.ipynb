```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
# 1. Load the pre-trained Haar Cascade for face detection
# This is the standard model provided by OpenCV to recognize facial features
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
# 2. List of your uploaded .webp images
image_list = ['img.webp', 'img2.webp', 'img3.webp', 'img4.webp']

print(f"--- Starting Face Detection for {len(image_list)} images ---")
```

```python

# 3. Process each image in a loop
for image_path in image_list:
    # Read the image from the file system
    img = cv2.imread(image_path)

    # Check if the image was actually loaded
    if img is None:
        print(f"Error: Could not find or load {image_path}")
        continue

    # Convert to Grayscale
    # Haar Cascades perform better and faster on single-channel (black and white) images
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces
    # scaleFactor:This tells the AI to look for faces of different sizes (some close to the camera, some far away).
    # minNeighbors:This ensures the AI is "sure" about a face by checking if several detection rectangles overlap in the same spot.
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around every detected face
    for (x, y, w, h) in faces:
        # cv2.rectangle(image, start_point, end_point, color_bgr, thickness)
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)

    # 4. Display the result for the current image
    print(f"\nResult for {image_path}: Found {len(faces)} face(s)")
    cv2_imshow(img)
```
